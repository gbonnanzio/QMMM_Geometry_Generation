```python
import MDAnalysis as mda
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Load the trajectory or PDB file
u = mda.Universe("5EJ5/5EJ5_crystal.pdb")

# Select protein and water oxygen atoms
protein = u.select_atoms("protein")
water_oxygens = u.select_atoms("resname HOH and name O")

# Initialize a list to store minimum distances
min_distances = []

# Calculate the minimum distance for each water oxygen to the protein
for water_oxygen in water_oxygens:
    distances = np.linalg.norm(protein.positions - water_oxygen.position, axis=1)
    min_distance = np.min(distances)
    min_distances.append(min_distance)

# Plot the distances in a histogram
plt.hist(min_distances, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Minimum Distances of Water Oxygens to Protein')
plt.xlabel('Distance (Å)')
plt.ylabel('Frequency')
plt.show()
```
